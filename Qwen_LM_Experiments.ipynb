#Qwen LM - a language model from Alibaba
Code based on
https://github.com/QwenLM/Qwen2.5

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# model_name = "Qwen/Qwen2.5-7B-Instruct"

model_name = "Qwen/Qwen2.5-0.5B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

prompt = "Give me a short introduction to large language models."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
print("response:", response)
```

#Conclusions

After conducting preliminary experiments with the Qwen2.5-0.5B model, which contains 0.5 billion parameters, the following conclusions can be drawn:

English Language: The model successfully generated a concise and coherent description of large language models (LLMs), demonstrating its capability in processing and articulating information in English.

Romanian Language: The results were less satisfactory, with numerous instances of hallucinations likely stemming from the model's inadequate training in Romanian. It created non-existent words and exhibited significant grammatical errors; however, it still managed to convey a somewhat meaningful introduction to LLMs despite these flaws.

Russian Language: While the model produced grammatically correct output, it diverged from the task of writing a straightforward introduction to LLMs. Instead, it focused on outlining various topics suitable for a presentation on LLMs, utilizing a scientific tone that, while sensible, did not align perfectly with the original assignment.

Overall, these findings highlight the strengths and limitations of the Qwen2.5-0.5B model across different languages. The model excels in English but faces challenges in Romanian and Russian, where it struggles with language-specific nuances and task adherence. Further research and targeted training could enhance its performance, particularly in underrepresented languages.
